```python
from ultralytics import YOLO
import os
```

```python
dataset_path = r"E:\submission\Part_1_Glove_Detection\dataset"
data_yaml = os.path.join(dataset_path, "data.yaml")
```

```python
assert os.path.exists(data_yaml), f"❌ data.yaml not found at {data_yaml}"
print("✅ Dataset and YAML file found!")
```

```python
import torch
print("CUDA available:", torch.cuda.is_available())
print("GPU name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU detected")
```

```python
model = YOLO("yolov8n.pt")  

results = model.train(
    data=data_yaml,
    epochs=2,                 
    imgsz=640,                
    batch=16,                  
    name="glove_detection",   
    project=r"E:\submission\Part_1_Glove_Detection\model", 
    workers=2,                 
    verbose=True
)

print("✅ Training complete!")

```
